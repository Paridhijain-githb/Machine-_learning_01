Separate X and Y features
Y : Species X : all remaining features, sepal length/width, petal length/width

```python
df.columns
```

```python
X = df.drop(columns=['price_range'])
Y = df[['price_range']]
```

```python
X.head()
```

```python
Y.head()
```

Data Preprocessing and Data Cleaning

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
pre = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
).set_output(transform='pandas')
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

Train Test Split
rate at which you can divide:

100%

60 % training 40% testing

70% training 30% testing

80% training 20% testing

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,train_size=0.7,test_size=0.3,random_state=21)
```

```python
xtrain.head()
```

```python
xtrain.index
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

Build the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.classes_
```

```python
ypred_train = model.predict(xtrain)
```

```python
yprob_train = model.predict_proba(xtrain)
```

```python
yprob_train[0]
```

```python
ytrain.head()
```

```python
xtrain.head()
```

Model Evaluation

```python
model.score(xtest,ytest)
```

```python
model.score(xtrain,ytrain)
```

Confusion Matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model,xtest,ytest)
```

Classification Report

```python
ypred_test = model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_test))
```

All the scores are greater than 80%(0.80), we can consider this model for final model building
