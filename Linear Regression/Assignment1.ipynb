Model Building : Linear Regression
Equation of Line: y = B0 + B1.X
B0 => y intercept
B1 => Slope of the line


ProfitPredicted = B0 + B1*YearsExperience 

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
#b0-y intercept
model.intercept_
```

```python
#b1-slope of the line
model.coef_
```

salary predicted = 24848.203 + 9449.9623*YearsExperience

```python
Y.head()
```

```python
Ypreds = model.predict(X)
Ypreds
```

```python
Ypreds[:5]
```

Evaluation Metrics
Mean Squared Error
Mean Absolute Error
RMSE
R2 squared- R2 score

```python

from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# Mean Squared Error
MSE = mean_squared_error(Y,Ypreds)
# RMSE - Root mean squared error
RMSE = MSE**(1/2)
# MAE 
MAE = mean_absolute_error(Y,Ypreds)
# R2 score
r2 = r2_score(Y,Ypreds)

print(f"Mean Squared Error: {MSE:.2f}")
print(f"Root Mean Squared Error: {RMSE:.2f}")
print(f"Mean Absolute Error: {MAE:.2f}")
print(f"R2 squared: {r2*100:.2f}%")
```

R2 squared value is greater than 80%, we can consider this model for final model building
